```python
import numpy as np
import pandas as pd
# from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score, f1_score
from sklearn.model_selection import train_test_split
import torch

from catboost import CatBoostClassifier, Pool

%matplotlib inline
     
```

```python
torch.cuda.is_available()
```

```python
train_df = pd.read_parquet('train_df_postprocess.pqt')   # Импортируем train и test
test_df = pd.read_parquet('test_df_postprocess.pqt')
```

```python
test_df.head(2)
```

## Model enginering

```python
cat_cols = []                         # Создаем список катерориальных признаков
for i in list(test_df.columns):       # Проходимся по каждому стробцу в test_df
    if test_df[i].dtypes == 'object': # Если столбец имеет тип 'object' добавляем его в список
        cat_cols.append(i)
```

```python
from catboost.utils import get_gpu_device_count
print(f'У вас {get_gpu_device_count()} GPU процессор')
```

### Модель предсказания start_cluster для month_6

```python

x1 = train_df.drop(["id_0", "end_cluster_0",'end_cluster_1','end_cluster_2',"start_cluster_2"], axis=1) # Убирем данные которые не должна видеть модель
y1 = train_df["start_cluster_2"]                          # Задаем целевую переменную

x_train, x_val, y_train, y_val = train_test_split(x1, y1,             # Разделяем данные на обучающие и валидационные данные
                                                  test_size=0.1,      # Размер валидационной выборки
                                                  random_state=42)    # Задаем random_state для воспроизводисмоти результата

cat_features =  [x_train.columns.get_loc(column) for column in cat_cols if column in x_train.columns] # Записываем индексы столбцов с категориальными признакоми

catboost = CatBoostClassifier(n_estimators=3000,           # Количество решающих деревьев
                              cat_features=cat_features,   # Передаем список с индексами категориальных признаков
                              depth = 8,                   # глубина решающих деревьев
                              task_type = 'GPU',           # Устройство на котором будет обучаться модель
                              random_seed =11,             # Задаем random seed для воспроизводимости результатов
                              l2_leaf_reg=2)               # (Значение по умоланию - 3) Уменьшим штрафы за большие веса из за малого количества сильно коррелирующих признаков с целевой переменной

catboost.fit(x_train, y_train)           #обучаем модель
model_start = catboost 
```

```python
y_pred = model_start.predict(x_val)
f1_score(y_pred,y_val,labels = model_start.classes_,average = 'micro')   # Выводим f1-меру для оценки качетсва модели
```

### Модель предсказания end_cluster для month_6

```python
X1 = train_df.drop(["id_0", "end_cluster_0",'end_cluster_1','end_cluster_2'], axis=1)
y1 = train_df["end_cluster_2"]

x_train, x_val, y_train, y_val = train_test_split(X1, y1,             # Разделяем данные на обучающие и валидационные данные
                                                  test_size=0.1,      # Размер валидационной выборки
                                                  random_state=42)    # Задаем random_state для воспроизводисмоти результата

cat_features =  [x_train.columns.get_loc(column) for column in cat_cols if column in x_train.columns] # Записываем индексы столбцов с категориальными признакоми

catboost = CatBoostClassifier(n_estimators=2000,             # Задаем количество решающих деревьев - 2000
                              cat_features=cat_features,     # Передаем список индексов категориальных признаков
                              depth = 7,                     # Устанавливаем глубину решающих деревьев - 7
                              task_type = 'GPU',             # Устройство для обучения модели
                              random_seed =11,               # Забаем random seed для воспроизводимости результатов
                              l2_leaf_reg=1.5)               # Уменьшаем значение Ridge из за малого количества сильно коррелирующих признаков с целевой переменной

catboost.fit(x_train, y_train)
model_end = catboost
```

```python
y_pred = model_end.predict(x_val)
f1_score(y_pred,y_val,labels = model_end.classes_,average = 'micro') # Выводим f1-меру для оценки качетсва модели
```

## Оценка качетва модели

```python
# Функция для определения roc auc метрики с учетом весов кластеров
def weighted_roc_auc(y_true, y_pred, labels, weights_dict):
    unnorm_weights = np.array([weights_dict[label] for label in labels]) # Создаем np array с помощью генератора списков (для кажого клсатера задаем вес)
    weights = unnorm_weights / unnorm_weights.sum()                      # Делим каждый вес клстеров на сумму всех весов кластеров
    classes_roc_auc = roc_auc_score(y_true, y_pred, labels=labels,       # Вычисляем roc-auc для каждого кластера 
                                    multi_class="ovr", average=None)
    return sum(weights * classes_roc_auc)                                # Возвращяем сумму (roc_auc для каждого кластера * на его вес)
```

```python
cluster_weights = pd.read_excel("./cluster_weights.xlsx").set_index("cluster")   # Читаем xlsx файл с весами для каждого кластера
weights_dict = cluster_weights["unnorm_weight"].to_dict()                # Создаем словарь {кластер:вес}
```

```python
y_pred_proba = model_end.predict_proba(x_val)  # Предсказываем на валидационных данных целевуб переменную (end_cluster_2) [_2 - для 6 месяца]
y_pred_proba.shape
```

```python
weighted_roc_auc(y_val, y_pred_proba, model_end.classes_, weights_dict) # Вызываем функцию для расчета roc-auc с учетом веса каждого кластера
```
