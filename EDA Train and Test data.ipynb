Исходя из Визуального просмотра и просмотра информации о наборах, можно сделать вывод:
* каждый id имеет временной промежуток с певого по третий месяц в тренеровном наборе и с 4 по 6 месяц в тесовом наборе.
* в тестовом наборе пропущены часть строк с 4 месяцем 'month_4'

Чтобы правильно расчитать процент пропусков тестовом наборе, нам нужно восстановить все строки с 4 месяцем.

```python
counts = test_df['id'].value_counts()  # массив id и кол-во записей с этим id
ids =  counts[counts<3].index.tolist() # нахождение id записей которых меньше 3 (т.е находим id тех самых пропущеных месяцев.)
new_rows = []
# МОДУЛЬ ДЛЯ ОБЬЕДИНЕНИЯ ДАННЫХ ПО id
for uid in ids:
        # Если записи нет, создаем новую строку с None для остальных колонок
    new_row = {'id': uid, 'date': 'month_4'}
    new_row.update({col: None for col in test_df.columns if col not in ['id', 'date']})
    new_rows.append(new_row)
new_rows_df = pd.DataFrame(new_rows)
test_df = pd.concat([test_df, new_rows_df], ignore_index=True)
test_df = test_df.sort_values(by=['id','date']).reset_index(drop=True)
# результат, востанновление данных id где отсутствовал 4 месяц, заполнятся None и Nan
```

 # Корреляция признаков между признаками 
 ----
 #### Так как признаков очень много, и визуализировать это, будет полная не разбериха, выбор страегии выпал на визуализацию только тех признаков коореляция которых более 0.05 .

```python
# Чтобы выявить корреляцию на нужно произвести кодировку категориальных признаков.
label_encoder = preprocessing.LabelEncoder() 

cat_feature = []
for col in train_df.columns:
    if train_df[col].dtype == 'object':
        cat_feature.append(col)
        

for feature in cat_feature: 
    train_df[feature]= label_encoder.fit_transform(train_df[feature]) 

train_df.head()
```

```python
correlations = train_df.corr(method='spearman')

# Нарисуем тепловую карту корреляций отфильтрованных признаков
plt.figure(figsize=(15, 10))  # Задаем размер фигуры для лучшей видимости
sns.heatmap(correlations, cmap='coolwarm')
plt.title("Тепловая карта корреляций отфильтрованных признаков")
plt.show()
```

## Визуализация целевой переменной, с самыми коррелируемыми признакми.

```python
# Вытаскиваем из таблицы коррелции, серию "end_cluster", 
# выбиаем только те признаки которые коррелируют  больше чем на 0.05
# Удаляем из серии корреляцию целевой переменной с самой собой же, это может сбить с толку, такак корреляция будет = 1.0
# Соритруем по убыванию, чтобы был порядок и приятнее для глаз
correlations_target = correlations[correlations['end_cluster']>0.05]['end_cluster'][:-1].sort_values(ascending=False)
```

```python
# Визуализация корреляций с целевой переменной
plt.figure(figsize=(10,6))
sns.barplot(x=correlations_target.index, y=correlations_target.values)
plt.title('Корреляция признаков с целевой переменной')
plt.ylabel('Коэффициент корреляции')
plt.xticks(rotation=70)

plt.show()
```
